**Line Charts**

```python
import pandas as pd
import numpy as np
unrate = pd.read_csv("data/civilian_unemployment_rate.csv")
unrate.columns
unrate.head()
```

```python
# Convert DATE column to pandas datetime object
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
```

```python
unrate.head(12) # January to December 1948
```

```python
# Matplotlib
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot()
plt.show()
```

```python
# Trends for 1948
# Fixing Axis Ticks
data1948 = unrate[:12]
plt.plot(data1948["DATE"],data1948["VALUE"])
plt.xticks(rotation=90) 
plt.show()
```

```python
# Adding Axis Labels and a Title
data1948 = unrate[:12]
plt.plot(data1948["DATE"],data1948["VALUE"])
plt.xticks(rotation=90) 
plt.xlabel('Month')
plt.ylabel("Unemployment Rate")
plt.title('Monthly Unemployment Trends, 1948')
plt.show()
```

**Multiple Plots**

```python
#  Matplotlib Classes
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
# Adding Data
ax1.plot(data1948["DATE"],data1948["VALUE"])
ax1.set_title('Monthly Unemployment Rate, 1948')
data1949 = unrate[12:24]
ax2.plot(data1949["DATE"],data1949["VALUE"])
ax2.set_title('Monthly Unemployment Rate, 1949')
# Formatting and Spacing
fig = plt.figure(figsize=(20,6)) # Regular way (Outside of Jupyter)
plt.rcParams["figure.figsize"] = (12,6) # Jupyter notebook way
```

```python
# comparison of multiple years
fig = plt.figure(figsize=(16,20))
ax1 = fig.add_subplot(5,1,1)
ax2 = fig.add_subplot(5,1,2)
ax3 = fig.add_subplot(5,1,3)
ax4 = fig.add_subplot(5,1,4)
ax5 = fig.add_subplot(5,1,5)

ax1.plot(unrate["DATE"][0*12:1*12],unrate["VALUE"][0*12:1*12])
ax2.plot(unrate["DATE"][1*12:2*12],unrate["VALUE"][1*12:2*12])
ax3.plot(unrate["DATE"][2*12:3*12],unrate["VALUE"][2*12:3*12])
ax4.plot(unrate["DATE"][3*12:4*12],unrate["VALUE"][3*12:4*12])
ax5.plot(unrate["DATE"][4*12:5*12],unrate["VALUE"][4*12:5*12])

ax1.set_title('Monthly Unemployment Rate, 1948')
ax2.set_title('Monthly Unemployment Rate, 1949')
ax3.set_title('Monthly Unemployment Rate, 1950')
ax4.set_title('Monthly Unemployment Rate, 1951')
ax5.set_title('Monthly Unemployment Rate, 1952')
```

```python
# Overlaying Line Charts
unrate["MONTH"] = unrate["DATE"].dt.month
fig = plt.figure(figsize=(6,3))
plt.plot(unrate["MONTH"][:12],unrate["VALUE"][:12],c='red')
plt.plot(unrate["MONTH"][12:24],unrate["VALUE"][12:24],c='blue')
```

```python
# Adding More Lines
fig = plt.figure(figsize=(10,6))
plt.plot(unrate["MONTH"][:12],unrate["VALUE"][:12],c='red')
plt.plot(unrate["MONTH"][1*12:2*12],unrate["VALUE"][1*12:2*12],c='blue')
plt.plot(unrate["MONTH"][2*12:3*12],unrate["VALUE"][2*12:3*12],c='green')
plt.plot(unrate["MONTH"][3*12:4*12],unrate["VALUE"][3*12:4*12],c='orange')
plt.plot(unrate["MONTH"][4*12:5*12],unrate["VALUE"][4*12:5*12],c='black')
```

```python
# Adding a Legend (and better way to plot multiple lines)
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
years = ['1948', '1949', '1950', '1951', '1952']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=years[i])

plt.legend(loc='upper left')
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel("Month, Integer")
plt.ylabel("Unempoyment Rate, Percent")
plt.show()
```
