```python
# 1: Introducing Thanksgiving Dinner Data
import pandas as pd
import numpy as np
data = pd.read_csv("data/thanksgiving.csv",encoding="Latin-1")
data.head(3)
```

```python
# 2: Filtering Out Rows forom a DataFrame
counts = pd.Series.value_counts(data['Do you celebrate Thanksgiving?'])
print(counts)
yesRows = data['Do you celebrate Thanksgiving?'] == 'Yes'
yesdata = data[yesRows]
newcounts = pd.Series.value_counts(yesdata['Do you celebrate Thanksgiving?'])
print(newcounts)
```

```python
# 3: Using Value_counts to Explore Main Dishes
dishCounts = pd.Series.value_counts(data['What is typically the main dish at your Thanksgiving dinner?'])
print(dishCounts)
tofurkeyRows = data['What is typically the main dish at your Thanksgiving dinner?'] == 'Tofurkey'
tofurkeyData = data[tofurkeyRows]
print(tofurkeyData)
print(tofurkeyData['Do you typically have gravy?'])
```

```python
# 4: Figuring Out What Pies People Eat
apple_isnull = pd.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple'])
pumpkin_isnull = pd.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin'])
pecan_isnull = pd.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan'])
ate_pies = data[apple_isnull & pumpkin_isnull & pecan_isnull]
ate_pies.head(2)
```

```python
# 5: Converting Age to Numeric
def stringToInt(s):
    if pd.isnull(s):
        return None
    else:
        s = s.split(' ')
        s = s[0]
        s = s.replace('+','')
        return int(s)
    
data['int_age'] = data['Age'].apply(lambda x: stringToInt(x))
data['int_age'].describe()
```

```python
# 6: Converting Income to Numeric
def moneyToInt(s):
    if pd.isnull(s):
        return None
    else:
        sl = s.split(' ')
        f = sl[0]
        f = f.replace(" ","")
        if f == "Prefer":
            return None
        else:
            f = f.replace("$"," ")
            f = f.replace(","," ")
            f = f.replace(" ","")
            return int(f)
                  

income = data["How much total combined money did all members of your HOUSEHOLD earn last year?"]
print(income.head(10))
data["int_income"] = income.apply(lambda x: moneyToInt(x))
data["int_income"].head(10)
```

```python
# 7: Correlating Travel Distance and Income
rows = data["int_income"] < 150000
under150k = data[rows]
under150k["How far will you travel for Thanksgiving?"].value_counts()
```

```python
# 8: Linking Friendship and Age

#print(data["int_age"])
notNullRows = data['int_age'] != None
newdata = data[notNullRows]


pv = newdata.pivot_table(index="Have you ever tried to meet up with hometown friends on Thanksgiving night?",columns="Have you ever attended a \"Friendsgiving?\"",values="int_age")
pv
```
